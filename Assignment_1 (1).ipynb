1.Using a for loop, print each job posting in a readable format.


```python
job_data = { "Job_ID": [1, 2, 3, 4, 5],
"Role": ["Data Analyst", "Data Scientist", "Business Analyst", "Data Engineer", "ML Engineer"],
"Company": ["ABC Corp", "XyZ Ltd", "DataWorks", "InfraTech", "AI Labs"],
"Skills": [
"Excel SQL Python",
"Python ML Statistics",
"Excel SQL PowerBI",
"Python SQL Spark",
"Python ML DeepLearning"
]}
print(job_data)
```

```python
num=len(job_data['Job_ID'])
for i in range(num):
  output=[]
  for key in job_data:
    if i<len(job_data[key]):
      output.append(f"{key}:{job_data[key][i]} |")
    else:
      output.append(f"{key}:0")
  print("".join(output))
```

2.Add New Job Postings (interactive)

```python
count=int(input("Enter the count"))
for i in range(count):
  output=[]
  role=input("Enter the role")
  company=input("Enter the company")
  skills=input("Enter the skills")
  job_data["Role"].append(role)
  job_data["Company"].append(company)
  job_data["Skills"].append(skills)
  job_data["Job_ID"].append(len(job_data["Job_ID"])+1)
num=len(job_data['Job_ID'])
for i in range(num):
  output=[]
  for key in job_data:
    if i<len(job_data[key]):
      output.append(f"{key}:{job_data[key][i]} |")
    else:
      output.append(f"{key}:0")
  print("".join(output))
```

3.Clean & Normalize Skills (text processing)

```python
def clean_skills():
  for i in range(len(job_data["Skills"])):
    job_data["Skills"][i]=job_data["Skills"][i].lower()
    job_data["Skills"][i]=job_data["Skills"][i].replace(","," ")
clean_skills()
print(job_data["Skills"])

```

4) Skill Extraction & Frequency

```python
known_skills = ['excel', 'sql', 'python', 'ml', 'statistics', 'powerbi', 'spark', 'deeplearning', 'data', 'decision', 'analyst', 'engineer']
skill_counts = {}
for skill in known_skills:
    skill_counts[skill] = 0
for job in job_data["Skills"]:
    individual_skills = job.lower().split()
    for sk in individual_skills:
        if sk in known_skills:
            skill_counts[sk] += 1
sorted_skills = sorted(skill_counts.items(), key=lambda item: item[1], reverse=True)
print("Top skills by frequency:")
count = 0
for skill, freq in sorted_skills:
    if freq > 0:
        print(f"{skill}: {freq}")
        count += 1
        if count >= 5:
            break
```

6) Skill Search Function (functions & conditional logic)

```python
def find_jobs_by_skill(skill_query):
  found = []
  for i in range(len(job_data["Skills"])):
    if skill_query in job_data["Skills"][i].lower():
      found.append(f"{job_data['Job_ID'][i]} | {job_data['Role'][i]} | {job_data['Company'][i]} | Skills: {job_data['Skills'][i]}")
  return found

s = input("Enter the skill").lower()
result = find_jobs_by_skill(s)
if len(result)==0:
  print("No jobs found")
else:
  for job in result:
    print(job)
```

```python
hig_freq=list(filter(lambda item: item[1] > 2, sorted_skills))
for skill,freq in hig_freq:
  print(f"{skill}:{freq}")
```
