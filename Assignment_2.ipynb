Task 3 – Create SaaS Customer Class (OOP)

```python
class SaasCustomer:
    def __init__(self, customer_id, name, signup_date, plan, monthly_fee, usage_hours):
        self.customer_id = customer_id
        self.name = name
        self.signup_date = signup_date
        self.plan = plan
        self.monthly_fee = monthly_fee
        self.usage_hours = usage_hours

    def validate_data(self):

        if self.name == None or self.name == "":
            self.name = "No Customer"

        if self.plan == None or self.plan == "":
            self.plan = "No plan"
        else:
            self.plan = self.plan.capitalize()

        if str(self.monthly_fee).replace(".", "").isdigit():
            self.monthly_fee = float(self.monthly_fee)
        else:
            self.monthly_fee = None

        if self.usage_hours == None or str(self.usage_hours).upper() == "NULL":
            self.usage_hours = 0
        else:
            if str(self.usage_hours).isdigit():
                self.usage_hours = int(self.usage_hours)
            else:
                self.usage_hours = 0

    def display_customer(self):
        print("Customer Name:", self.name)
```

Task 4 – Data Cleaning Pipeline

```python
from datetime import datetime
def handle_missing_values(data):
    n = len(data["customer_id"])
    for i in range(n):
        if data["name"][i] == None or data["name"][i] == "":
            data["name"][i] = "No Customer"

        if data["usage_hours"][i] == None or str(data["usage_hours"][i]).upper() == "NULL":
            data["usage_hours"][i] = 0
        else:
            if str(data["usage_hours"][i]).isdigit():
                data["usage_hours"][i] = int(data["usage_hours"][i])
            else:
                data["usage_hours"][i] = 0

        fee = str(data["monthly_fee"][i]).replace(".", "", 1)
        if fee.isdigit():
            data["monthly_fee"][i] = float(data["monthly_fee"][i])
        else:
            data["monthly_fee"][i] = None

        if data["plan"][i] == None or data["plan"][i] == "":
            data["plan"][i] = "No Plan"
        else:
            data["plan"][i] = data["plan"][i].capitalize()
    return data


def remove_duplicates(data):
    seen = set()
    new_data = {
        "customer_id": [],
        "name": [],
        "signup_date": [],
        "plan": [],
        "monthly_fee": [],
        "usage_hours": []
    }

    n = len(data["customer_id"])

    for i in range(n):
        new_id = data["customer_id"][i]

        if new_id not in seen:
            seen.add(new_id)

            for key in data:
                new_data[key].append(data[key][i])

    return new_data

saas_data = handle_missing_values(saas_data)
saas_data = remove_duplicates(saas_data)

n = len(saas_data["customer_id"])
for num in range(n):
    output = []
    for key in saas_data:
        output.append(f"{key}:{saas_data[key][num]} | ")
    print("".join(output))
```

Task 5 – Summary Statistics and Insights

```python
def summary_statistics(data):
    total_customers = len(data["customer_id"])
    total_fee = 0
    fee_count = 0
    total_usage = 0
    plan_count = {}
    for i in range(total_customers):
        if data["monthly_fee"][i] != None:
            total_fee += data["monthly_fee"][i]
            fee_count += 1

        total_usage += data["usage_hours"][i]

        plan = data["plan"][i]
        if plan in plan_count:
            plan_count[plan] += 1
        else:
            plan_count[plan] = 1

    avg_fee = 0
    if fee_count != 0:
        avg_fee = total_fee / fee_count

    print("Total Customers:", total_customers)
    print("Average Monthly Fee:", round(avg_fee, 2))
    print("Total Usage Hours:", total_usage)
    print("\nPlan Distribution")

    for p in plan_count:
        print(p, "-", plan_count[p])

summary_statistics(saas_data)
```

Task 6 – Identify Low-Usage Customers

```python
def find_low_usage_customers(data):
    print("Low Usage Customers\n")
    n = len(data["customer_id"])
    for i in range(n):
        if data["usage_hours"][i] < 50:
            print("ID:", data["customer_id"][i], "|",
                  "Name:", data["name"][i], "|",
                  "Plan:", data["plan"][i], "|",
                  "Usage Hours:", data["usage_hours"][i])
find_low_usage_customers(saas_data)
```

Task 8 – Save Cleaned Data to File

```python
def save_cleaned_data(data):
    try:
        file = open("saas_customers_cleaned.csv", "w")
        file.write("customer_id,name,signup_date,plan,monthly_fee,usage_hours\n")
        n = len(data["customer_id"])
        for i in range(n):
            row = f"{data['customer_id'][i]},{data['name'][i]},{data['signup_date'][i]},{data['plan'][i]},{data['monthly_fee'][i]},{data['usage_hours'][i]}\n"
            file.write(row)
    except:
        print("Error ")
    else:
        print("created")
    finally:
        file.close()


def summary(data):
    try:
        plan_count = {}
        n = len(data["plan"])
        for i in range(n):
            plan = data["plan"][i]
            if plan in plan_count:
                plan_count[plan] += 1
            else:
                plan_count[plan] = 1
        file = open("plan_summary.txt", "w")
        for plan in plan_count:
            file.write(f"{plan} : {plan_count[plan]}\n")
    except:
        print("Error")
    else:
        print("created")
    finally:
        file.close()

save_cleaned_data(saas_data)
summary(saas_data)
```
